```python
#get file with trial-by-trial responses
responses_dir = os.path.join(acquisition_dir, opts.combined_run,'responses',traceid)
data_array_dir = os.path.join(responses_dir, 'data_arrays')

resp_array_fn = 'trial_response_array.hdf5'
resp_array_filepath = os.path.join(data_array_dir,'files' , resp_array_fn)
resp_grp = h5py.File(resp_array_filepath, 'r')

if 's2p_cell_rois' in resp_grp.attrs.keys():
    cell_rois = resp_grp.attrs['s2p_cell_rois']
else:
    cell_rois = np.arange(nrois)

curr_slice = 'Slice01'#hard,coding for now

#unpack
response_matrix = np.array(resp_grp['/'.join([curr_slice, 'responses' ,response_type])])
response_matrix = response_matrix[:,:,cell_rois]
ntrials,nconfigs,nrois = response_matrix.shape

filter_crit_matrix_trials = np.array(resp_grp['/'.join([curr_slice, 'responses' ,filter_crit])])
filter_crit_matrix_trials = filter_crit_matrix_trials[:,:,cell_rois]
resp_grp.close()


filter_crit_matrix_mean = np.squeeze(np.mean(filter_crit_matrix_trials,0))

#consider a config active if at least one of theversion of an image evoked a response above threshol
thresh = filter_thresh
thresh_matrix = filter_crit_matrix_mean>thresh
filter_matrix = np.ones((thresh_matrix.shape))*np.nan
for ridx in range(nrois):
    for idx in range(0,thresh_matrix.shape[0],3):
            if np.sum(thresh_matrix[idx:idx+3,ridx])>0:
                filter_matrix[idx:idx+3,ridx] = 1



#load trial info
stimconfig_fn = 'trial_conditions.hdf5'
stimconfig_filepath = os.path.join(paradigm_dir, 'files', stimconfig_fn)
config_grp = h5py.File(stimconfig_filepath, 'r')

trial_config = np.array(config_grp['trial_config'])
trial_cond = np.array(config_grp['trial_cond'])
trial_img = np.array(config_grp['trial_img'])

config_grp.close()

#for each roi, first average across trials per config, then normalize across configs
#-should result in higher values, also noticed bigger differences in values

nconfigs = len(np.unique(trial_config))
print('ROIs:%i'%(nrois))
print('Trials:%i'%(ntrials))
```

```python
mean_response_matrix = np.mean(response_matrix,0)
norm_response_array = np.empty((nconfigs,nrois))

for ridx in range(nrois):
    norm_response_array[:,ridx] = mean_response_matrix[:,ridx]/np.nanmax(response_matrix[:,ridx])
    
filt_norm_response_array = norm_response_array*filter_matrix
```

```python
if 's2p_cell_rois' in resp_grp.attrs.keys():
    cell_rois = resp_grp.attrs['s2p_cell_rois']
else:
    cell_rois = np.arange(nrois)
config_img = resp_grp.attrs['config_img'].astype('int')
config_cond = resp_grp.attrs['config_cond'].astype('int')

curr_slice = 'Slice01'#hard,coding for now
```

```python
#unpack
response_matrix_df_f = np.array(resp_grp['/'.join([curr_slice, 'responses' ,'df_f'])])
response_matrix_zscore = np.array(resp_grp['/'.join([curr_slice, 'responses' ,'zscore'])])
```

```python
print(response_matrix_df_f.shape)
```

```python
#get average response across trials, for cell ROIs only
mean_df_f = np.mean(response_matrix_df_f[:,:,cell_rois],0)
se_df_f = stats.sem(response_matrix_df_f[:,:,cell_rois],0)
mean_zscore= np.mean(response_matrix_zscore[:,:,cell_rois],0)
se_zscore = stats.sem(response_matrix_zscore[:,:,cell_rois],0)
```

```python
#get number of active cells - (threshold by z-score)
nrois = len(cell_rois)
valid_responses = np.zeros((len(config_img),nrois))#to keep track of configs where cell responss above thresh
active_roi_count = 0
active_roi_list = []
thresh = 1 #using this for now, should verify with split-halves analysis
for ridx in range(nrois):    
    if np.any(mean_zscore[:,ridx]>thresh):
        active_roi_list.append(cell_rois[ridx])
    for img in np.unique(config_img):
        #img=16
        config_idxs = np.where(config_img==img)[0]
        if np.any(mean_zscore[config_idxs,ridx]>thresh):
            valid_responses[config_idxs,ridx] = 1
        
active_roi_count = len(active_roi_list)     
print('%d out of %d cell active '%(active_roi_count,nrois))
print('Fraction ROIs active: %.04f'%(float(active_roi_count)/nrois))
```

```python
resp_diff = np.zeros(mean_df_f.shape)
mod_idx = np.zeros(mean_df_f.shape)
for ridx in range(nrois): 
    
    for cond_idx in range(0,config_img.size,3):
        if valid_responses[cond_idx,ridx]:
            #tex - sf
            resp_diff[cond_idx,ridx] = mean_df_f[cond_idx+2,ridx] - mean_df_f[cond_idx+1,ridx]
            mod_idx[cond_idx,ridx] = np.true_divide(mean_df_f[cond_idx+2,ridx] - mean_df_f[cond_idx+1,ridx],\
                                                    mean_df_f[cond_idx+2,ridx] + mean_df_f[cond_idx+1,ridx])
            #orig - sf
            resp_diff[cond_idx+1,ridx] = mean_df_f[cond_idx,ridx] - mean_df_f[cond_idx+1,ridx]
            mod_idx[cond_idx+1,ridx] = np.true_divide(mean_df_f[cond_idx,ridx] - mean_df_f[cond_idx+1,ridx],\
                                                    mean_df_f[cond_idx,ridx] + mean_df_f[cond_idx+1,ridx])
            #orig - tex
            resp_diff[cond_idx+2,ridx] = mean_df_f[cond_idx,ridx] - mean_df_f[cond_idx+2,ridx]
            mod_idx[cond_idx+2,ridx] = np.true_divide(mean_df_f[cond_idx,ridx] - mean_df_f[cond_idx+2,ridx],\
                                                    mean_df_f[cond_idx,ridx] + mean_df_f[cond_idx+2,ridx])
        else:
            resp_diff[cond_idx:cond_idx+3,ridx] = np.nan
            mod_idx[cond_idx:cond_idx+3,ridx] = np.nan
```

```python
nvalid_cells = np.sum(valid_responses,1).astype('int')
#averaging response across active neurons (1 active condition) - group by image

#set non-responsive image responses to nan
invalid_idx = np.where(valid_responses==0)
mean_df_f_valid = np.copy(mean_df_f)
mean_df_f_valid[invalid_idx[0],invalid_idx[1]]= np.nan
```

```python
img_df_f_mean = np.nanmean(mean_df_f_valid,1)
img_df_f_se = np.true_divide(np.nanstd(mean_df_f_valid,1),np.sqrt(nvalid_cells))
```
